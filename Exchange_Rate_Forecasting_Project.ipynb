```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100


def evaluate(y_true, y_pred):
    # Align the true and predicted values based on index and drop NaNs
    comparison_df = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred}).dropna()
    y_true_aligned = comparison_df['y_true']
    y_pred_aligned = comparison_df['y_pred']

    return {
        'RMSE': np.sqrt(mean_squared_error(y_true_aligned, y_pred_aligned)),
        'MAE': mean_absolute_error(y_true_aligned, y_pred_aligned),
        'MAPE': mape(y_true_aligned, y_pred_aligned)
    }
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Perform Augmented Dickey-Fuller test
adf_result = adfuller(series['Rate'].dropna())
print(f"ADF Statistic: {adf_result[0]:.3f}")
print(f"p-value: {adf_result[1]:.3f}")

# Plot ACF and PACF
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(series['Rate'].dropna(), ax=axes[0], lags=50)
plot_pacf(series['Rate'].dropna(), ax=axes[1], lags=50)
plt.show()
```

## **Prophet Model**

```python
from prophet import Prophet

# Prepare data for Prophet (requires 'ds' and 'y' columns)
prophet_data = series.reset_index().rename(columns={'Time Serie': 'ds', 'Rate': 'y'})

# Split data
train_size = int(len(prophet_data) * 0.8)
train_prophet = prophet_data.iloc[0:train_size]
test_prophet = prophet_data.iloc[train_size:]

# Initialize and fit Prophet model
model_prophet = Prophet()
model_prophet.fit(train_prophet.dropna()) # Drop NaNs for training

# Make future dataframe for prediction
future = model_prophet.make_future_dataframe(periods=len(test_prophet))

# Make predictions
forecast = model_prophet.predict(future)

# Merge test data with forecast to align predictions with actuals
merged_test_forecast = pd.merge(test_prophet, forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], on='ds', how='inner')

# Extract actual and predicted values for evaluation
y_true_prophet = merged_test_forecast['y']
y_pred_prophet = merged_test_forecast['yhat']


# Evaluate Prophet model
metrics_prophet = evaluate(y_true_prophet.dropna(), y_pred_prophet.dropna())
print("\nProphet Model Evaluation:")
print(metrics_prophet)

# Plot the forecast
fig = model_prophet.plot(forecast)
plt.title('Prophet Forecast')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.show()

# Plot components
fig2 = model_prophet.plot_components(forecast)
plt.show()
```

## **XGBOOST**

```python
# Prepare data for XGBoost (Feature Engineering)
# Create lag features and rolling window features
def create_features(df, label=None):
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.isocalendar().week.astype(int)

    # Add lag features
    for lag in range(1, 8):
        df[f'Rate_Lag_{lag}'] = df['Rate'].shift(lag)

    # Add rolling window features
    df['Rate_Rolling_Mean_7'] = df['Rate'].rolling(window=7).mean()
    df['Rate_Rolling_Std_7'] = df['Rate'].rolling(window=7).std()

    if label:
        df[label] = df[label].copy()

    return df

series_features = create_features(series.copy(), 'Rate')

# Drop rows with NaN values created by feature engineering
series_features.dropna(inplace=True)

# Split data into training and testing sets
train_size_ml = int(len(series_features) * 0.8)
train_ml = series_features.iloc[0:train_size_ml]
test_ml = series_features.iloc[train_size_ml:]

# Define features (X) and target (y)
features = [col for col in series_features.columns if col not in ['date', 'Rate']]
target = 'Rate'

X_train = train_ml[features]
y_train = train_ml[target]
X_test = test_ml[features]
y_test = test_ml[target]

# Initialize and train XGBoost Regressor
xgbr = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000, learning_rate=0.01, n_jobs=-1)
xgbr.fit(X_train, y_train)

# Make predictions
xgb_pred = xgbr.predict(X_test)

# Evaluate XGBoost model
metrics_xgboost = evaluate(y_test, xgb_pred)
print("\nXGBoost Model Evaluation:")
print(metrics_xgboost)

# Plot XGBoost forecast
plt.figure(figsize=(10, 4))
plt.plot(y_train.index, y_train, label='Train')
plt.plot(y_test.index, y_test, label='Test')
plt.plot(y_test.index, xgb_pred, label='XGBoost Forecast')
plt.title('XGBoost Forecast vs Actuals')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')
plt.legend(); plt.show()
```

Build and train the LSTM model.

**Reasoning**:
Prepare the data for the LSTM model by selecting the 'Rate' column, handling missing values, scaling, creating sequences, converting to numpy arrays, reshaping for LSTM, and splitting into train and test sets.

```python
# 1. Select the 'Rate' column
rate_series = series['Rate']

# 2. Handle missing values
rate_series = rate_series.ffill().bfill()

# 3. Scale the 'Rate' series
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(rate_series.values.reshape(-1, 1))

# 4. Create sequences for the LSTM model
def create_sequences(data, look_back=1):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back), 0])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

look_back = 60
X, y = create_sequences(scaled_data, look_back)

# 6. Convert to NumPy arrays (already done by create_sequences)

# 7. Reshape input sequences for LSTM
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# 8. Split into training and testing sets
# Adjust the split point based on the look_back value
train_size_lstm = train_size - look_back
X_train_lstm, X_test_lstm = X[0:train_size_lstm,:], X[train_size_lstm:len(X),:]
y_train_lstm, y_test_lstm = y[0:train_size_lstm], y[train_size_lstm:len(y)]

print("X_train_lstm shape:", X_train_lstm.shape)
print("y_train_lstm shape:", y_train_lstm.shape)
print("X_test_lstm shape:", X_test_lstm.shape)
print("y_test_lstm shape:", y_test_lstm.shape)
```

```python
# 1. Import necessary layers and models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 2. Instantiate a Sequential model
model_lstm = Sequential()

# 3. Add an LSTM layer
model_lstm.add(LSTM(units=50, return_sequences=False, input_shape=(X_train_lstm.shape[1], 1)))

# 4. Add a Dense output layer
model_lstm.add(Dense(units=1))

# 5. Compile the model
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# 6. Train the LSTM model
history = model_lstm.fit(X_train_lstm, y_train_lstm, epochs=100, batch_size=32)

print("LSTM model training complete.")
```

## Consolidated Code for Streamlit Dashboard

Below is the consolidated Python code from this notebook, ready to be copied into a `.py` file for your Streamlit dashboard.

```python
import pandas as pd
import numpy as np
from datetime import datetime
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from arch import arch_model  # not used but kept
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import torch  # not used but kept
import xgboost as xgb
from catboost import CatBoostRegressor
import matplotlib.pyplot as plt
import streamlit as st
import warnings

warnings.filterwarnings('ignore')

# --- Data Loading and Preprocessing ---

# Load the dataset
try:
    df = pd.read_excel('Foreign_Exchange_Rates[1].xlsx')
except FileNotFoundError:
    st.error("Error: 'Foreign_Exchange_Rates[1].xlsx' not found. Please ensure the file is in the same directory as your Streamlit app.")
    st.stop()

# Select a currency exchange rate series for analysis
series = df[['Time Serie', 'EURO AREA - EURO/US$']].copy()
series['EURO AREA - EURO/US$'] = pd.to_numeric(series['EURO AREA - EURO/US$'], errors='coerce')

# Set the 'Time Serie' column as the index and ensure it's datetime
series['Time Serie'] = pd.to_datetime(series['Time Serie'])
series = series.set_index('Time Serie')
series = series.rename(columns={'EURO AREA - EURO/US$': 'Rate'})

# Handle missing values
series['Rate'] = series['Rate'].ffill().bfill()

# --- Feature Engineering (for ML models) ---
def create_features(df, label=None):
    df = df.copy()
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.isocalendar().week.astype(int)

    # Add lag features
    for lag in range(1, 8):
        df[f'Rate_Lag_{lag}'] = df['Rate'].shift(lag)

    # Add rolling window features
    df['Rate_Rolling_Mean_7'] = df['Rate'].rolling(window=7).mean()
    df['Rate_Rolling_Std_7'] = df['Rate'].rolling(window=7).std()

    if label:
        df[label] = df[label].copy()

    return df

series_features = create_features(series.copy(), 'Rate')
series_features.dropna(inplace=True)

# --- Data Splitting ---
train_size = int(len(series) * 0.8)
train = series.iloc[0:train_size]
test = series.iloc[train_size:]

train_size_ml = int(len(series_features) * 0.8)
train_ml = series_features.iloc[0:train_size_ml]
test_ml = series_features.iloc[train_size_ml:]

# LSTM dataset creation
look_back = 60
scaled_data = MinMaxScaler(feature_range=(0, 1)).fit_transform(series['Rate'].values.reshape(-1, 1))

def create_sequences(data, look_back=1):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back), 0])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data, look_back)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Adjust train/test split for sequences
train_size_lstm = train_size - look_back
X_train_lstm, X_test_lstm = X[0:train_size_lstm, :], X[train_size_lstm:len(X), :]
y_train_lstm, y_test_lstm = y[0:train_size_lstm], y[train_size_lstm:len(y)]
scaler_lstm = MinMaxScaler(feature_range=(0, 1)).fit(series['Rate'].values.reshape(-1, 1))  # For inverse transform

# --- Evaluation Metrics ---
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def evaluate(y_true, y_pred):
    comparison_df = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred}).dropna()
    y_true_aligned = comparison_df['y_true']
    y_pred_aligned = comparison_df['y_pred']

    if y_true_aligned.empty:
        return {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    return {
        'RMSE': np.sqrt(mean_squared_error(y_true_aligned, y_pred_aligned)),
        'MAE': mean_absolute_error(y_true_aligned, y_pred_aligned),
        'MAPE': mape(y_true_aligned, y_pred_aligned)
    }

# --- Model Training and Prediction ---
@st.cache_resource
def train_models(train_data, test_data, train_ml_data, test_ml_data, X_train_lstm_data, y_train_lstm_data, X_test_lstm_data, y_test_lstm_data, series_full, train_size_param):
    # ARIMA
    try:
        arima = ARIMA(train_data['Rate'].dropna(), order=(5, 1, 0)).fit()
        arima_pred = arima.forecast(steps=len(test_data))
        arima_pred = pd.Series(arima_pred, index=test_data.index[:len(arima_pred)])
        metrics_arima = evaluate(test_data['Rate'].dropna().loc[arima_pred.index], arima_pred)
    except Exception as e:
        st.warning(f"ARIMA model training or evaluation failed: {e}")
        arima_pred = pd.Series(index=test_data.index)
        metrics_arima = {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    # SARIMA
    try:
        sarima_model = SARIMAX(train_data['Rate'].dropna(), order=(1, 1, 1), seasonal_order=(1, 1, 1, 30), enforce_stationarity=False, enforce_invertibility=False)
        sarima_fit = sarima_model.fit(disp=False)
        sarima_pred = sarima_fit.forecast(steps=len(test_data))
        sarima_pred = pd.Series(sarima_pred, index=test_data.index[:len(sarima_pred)])
        metrics_sarima = evaluate(test_data['Rate'].dropna().loc[sarima_pred.index], sarima_pred)
    except Exception as e:
        st.warning(f"SARIMA model training or evaluation failed: {e}")
        sarima_pred = pd.Series(index=test_data.index)
        metrics_sarima = {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    # Prophet
    try:
        prophet_data_train = train_data.reset_index().rename(columns={'Time Serie': 'ds', 'Rate': 'y'})
        # If the column is not named 'Time Serie' after reset, ensure it gets named 'ds'
        if 'ds' not in prophet_data_train.columns:
            # take the first column (the reset index) and rename it to 'ds'
            first_col = prophet_data_train.columns[0]
            prophet_data_train = prophet_data_train.rename(columns={first_col: 'ds'})
        model_prophet = Prophet()
        model_prophet.fit(prophet_data_train.dropna())
        future = model_prophet.make_future_dataframe(periods=len(test_data), freq=None)
        forecast = model_prophet.predict(future)
        prophet_pred = forecast.set_index('ds')['yhat'].reindex(test_data.index)  # align to test index
        prophet_pred = prophet_pred.fillna(method='ffill').fillna(method='bfill')[:len(test_data)]
        metrics_prophet = evaluate(test_data['Rate'].dropna().loc[prophet_pred.dropna().index], prophet_pred.dropna())
    except Exception as e:
        st.warning(f"Prophet model training or evaluation failed: {e}")
        prophet_pred = pd.Series(index=test_data.index)
        metrics_prophet = {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    # XGBoost
    features_ml = [col for col in train_ml_data.columns if col not in ['date', 'Rate']]
    target_ml = 'Rate'
    X_train_ml = train_ml_data[features_ml]
    y_train_ml = train_ml_data[target_ml]
    X_test_ml = test_ml_data[features_ml]
    y_test_ml = test_ml_data[target_ml]
    try:
        xgbr = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000, learning_rate=0.01, n_jobs=-1)
        xgbr.fit(X_train_ml, y_train_ml)
        xgb_pred = xgbr.predict(X_test_ml)
        xgb_pred = pd.Series(xgb_pred, index=y_test_ml.index)
        metrics_xgboost = evaluate(y_test_ml, xgb_pred)
    except Exception as e:
        st.warning(f"XGBoost model training or evaluation failed: {e}")
        xgb_pred = pd.Series(index=y_test_ml.index)
        metrics_xgboost = {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    # LSTM
    try:
        model_lstm = Sequential()
        model_lstm.add(LSTM(units=50, return_sequences=False, input_shape=(X_train_lstm_data.shape[1], 1)))
        model_lstm.add(Dense(units=1))
        model_lstm.compile(optimizer='adam', loss='mean_squared_error')
        model_lstm.fit(X_train_lstm_data, y_train_lstm_data, epochs=50, batch_size=32, verbose=0)

        lstm_pred_scaled = model_lstm.predict(X_test_lstm_data)
        lstm_pred = scaler_lstm.inverse_transform(lstm_pred_scaled)

        # Align LSTM predictions to the test index properly
        lstm_test_index = test_data.index[:len(lstm_pred)]
        lstm_pred_series = pd.Series(lstm_pred.flatten(), index=lstm_test_index)

        y_test_lstm_actual = scaler_lstm.inverse_transform(y_test_lstm_data.reshape(-1, 1))
        y_test_lstm_series = pd.Series(y_test_lstm_actual.flatten(), index=lstm_test_index)

        metrics_lstm = evaluate(y_test_lstm_series, lstm_pred_series)
    except Exception as e:
        st.warning(f"LSTM model training or evaluation failed: {e}")
        lstm_pred_series = pd.Series(index=test_data.index[:len(y_test_lstm_data)])
        y_test_lstm_series = pd.Series(index=test_data.index[:len(y_test_lstm_data)])
        metrics_lstm = {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    # CatBoost
    features_catboost = [col for col in train_ml_data.columns if col not in ['date', 'Rate']]
    target_catboost = 'Rate'
    X_train_catboost = train_ml_data[features_catboost]
    y_train_catboost = train_ml_data[target_catboost]
    X_test_catboost = test_ml_data[features_catboost]
    y_test_catboost = test_ml_data[target_catboost]
    try:
        catboost_model = CatBoostRegressor(iterations=500, learning_rate=0.01, depth=6, loss_function='RMSE', eval_metric='RMSE', verbose=0, random_seed=42)
        catboost_model.fit(X_train_catboost, y_train_catboost, eval_set=(X_test_catboost, y_test_catboost), early_stopping_rounds=50, use_best_model=True)
        catboost_pred = catboost_model.predict(X_test_catboost)
        catboost_pred = pd.Series(catboost_pred, index=y_test_catboost.index)
        metrics_catboost = evaluate(y_test_catboost, catboost_pred)
    except Exception as e:
        st.warning(f"CatBoost model training or evaluation failed: {e}")
        catboost_pred = pd.Series(index=y_test_catboost.index)
        metrics_catboost = {'RMSE': np.nan, 'MAE': np.nan, 'MAPE': np.nan}

    return (metrics_arima, arima_pred,
            metrics_sarima, sarima_pred,
            metrics_prophet, prophet_pred,
            metrics_xgboost, xgb_pred,
            metrics_lstm, lstm_pred_series, y_test_lstm_series,
            metrics_catboost, catboost_pred,
            y_test_ml)  # return y_test_ml (actual targets for ML models)

# --- Run Models ---
results = train_models(train, test, train_ml, test_ml, X_train_lstm, y_train_lstm, X_test_lstm, y_test_lstm, series, train_size)
(metrics_arima, arima_pred,
 metrics_sarima, sarima_pred,
 metrics_prophet, prophet_pred,
 metrics_xgboost, xgb_pred,
 metrics_lstm, lstm_pred_series, y_test_lstm_series,
 metrics_catboost, catboost_pred,
 y_test_ml_actual) = results

# --- Dashboard Components ---
st.title("Currency Exchange Rate Forecasting Dashboard")

st.write("## Model Performance Comparison")
model_performance = {
    'ARIMA': metrics_arima,
    'SARIMA': metrics_sarima,
    'Prophet': metrics_prophet,
    'XGBoost': metrics_xgboost,
    'LSTM': metrics_lstm,
    'CatBoost': metrics_catboost
}
performance_df = pd.DataFrame.from_dict(model_performance, orient='index')
st.write("### Evaluation Metrics (Lower is Better)")
st.dataframe(performance_df)

st.write("### Performance Chart")
fig_perf, ax_perf = plt.subplots(figsize=(10, 6))
# Some metrics may be NaN; select available columns
cols_to_plot = [c for c in ['RMSE', 'MAE'] if c in performance_df.columns]
if not performance_df[cols_to_plot].dropna(how='all').empty:
    performance_df[cols_to_plot].plot(kind='bar', ax=ax_perf)
ax_perf.set_title('Model Performance Comparison (RMSE & MAE)')
ax_perf.set_ylabel('Error')
ax_perf.tick_params(axis='x', rotation=45)
ax_perf.legend(loc='upper right')
st.pyplot(fig_perf)

# Forecast Visualization
st.write("## Model Forecasts vs Actuals")
fig_forecast, ax_forecast = plt.subplots(figsize=(14, 7))
ax_forecast.plot(test.index, test['Rate'], label='Actual')

if not arima_pred.empty:
    ax_forecast.plot(arima_pred.index, arima_pred, label='ARIMA Forecast', linestyle='--')
if not sarima_pred.empty:
    ax_forecast.plot(sarima_pred.index, sarima_pred, label='SARIMA Forecast', linestyle='--')
if not prophet_pred.empty:
    ax_forecast.plot(prophet_pred.index, prophet_pred, label='Prophet Forecast', linestyle='--')
if not xgb_pred.empty:
    ax_forecast.plot(xgb_pred.index, xgb_pred, label='XGBoost Forecast', linestyle='--')
if not lstm_pred_series.empty:
    ax_forecast.plot(lstm_pred_series.index, lstm_pred_series, label='LSTM Forecast', linestyle='--')
if not catboost_pred.empty:
    ax_forecast.plot(catboost_pred.index, catboost_pred, label='CatBoost Forecast', linestyle='--')

ax_forecast.set_title('Model Forecasts vs Actuals')
ax_forecast.set_xlabel('Time')
ax_forecast.set_ylabel('Exchange Rate')
ax_forecast.legend()
ax_forecast.grid(True)

st.pyplot(fig_forecast)

st.write("## Conclusion")
st.write("""
This project aimed to develop a robust framework for forecasting currency exchange rates using several methods. The code runs multiple models and visualizes their forecasts and metrics. Make sure to further tune hyperparameters and confirm Prophet's frequency aligns with your data if you want the Prophet forecasts to be exact.
""")
```
